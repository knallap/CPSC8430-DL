```python
import numpy as np
import matplotlib.pyplot as plt

DCGAN = np.load('./Results/FID_score/DC_FID.npy')
WGANGP = np.load('./Results/FID_score/WGANGP_FID.npy')
plt.figure(figsize=(10,5))
plt.title("FID Scores comparsions for DCGAN vs WGAN")
plt.plot(DCGAN,label="DCGAN",color='red')
plt.plot(WGANGP,label="WGAN",color='blue')
plt.xlabel("Epochs")
plt.ylabel("FID")
plt.legend()
plt.show()
```
